```python
def clean_companyboards(companyboards):
    clist = companyboards.split(',')
    clist = [company.strip() for company in clist]
    seen = set()
    unique_clist = []
    for company in clist:
        if company not in seen:
            seen.add(company)
            unique_clist.append(company)
    return ', '.join(unique_clist)

nw_df['companyboards'] = nw_df['companyboards'].apply(clean_companyboards)
#nw_df
```

In this notebook I will create a new dataframe to make a network graph. I will have to make different attributes

```python
import requests
import pandas as pd
```

```python
groups_df = pd.read_csv("seats.csv")
# groups_df.head(30)
```

```python
tidy_data = (
    groups_df
    .assign(companyboards=groups_df['companyboards'].str.split(', '))
    .explode('companyboards')
    .rename(columns={'personname': 'source', 'companyboards': 'target', 'company_count': 'value'})
    .reset_index(drop=True)
)

# tidy_data.head(50)
```

```python
# I noticed that there was still one company wrong from cleaning data wrong

tidy_data["target"] = (
    tidy_data["target"]
    .str.replace("NY", "NYAB")
)
tidy_data
```

```python
csv_file_path = 'network.csv'
tidy_data.to_csv(csv_file_path, index=False)
from IPython.display import FileLink
FileLink(csv_file_path)
```
